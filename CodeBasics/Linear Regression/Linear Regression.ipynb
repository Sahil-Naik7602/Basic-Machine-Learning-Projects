Where **price** is a *dependent variable* and **area** is an *independent variable*.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("houseprice.csv")
df
```

```python
%matplotlib inline
plt.xlabel("area(in sq ft)")
plt.ylabel("price(in US$)")
plt.scatter(df.area,df.price,marker = '+',color = 'red')
```

```python
reg = linear_model.LinearRegression()
x_input = df[['area']]
y_actual = df['price']
reg.fit(x_input,y_actual)
```

```python
## Prediction
reg.predict([[3300]])
```

```python
##Slope of line
reg.coef_
```

```python
##Intercept Of line
reg.intercept_
```

```python
d = pd.read_csv('area.csv')
```

```python
d['prices'] = reg.predict(d)
```

```python
d
```

```python
d.to_csv("prediction_Of_area.csv",index = False)
```

```python
##Finding the best fit line of Regression
%matplotlib inline
plt.xlabel("area(in sq ft)")
plt.ylabel("price(in US$)")
plt.scatter(df.area,df.price,marker = '+',color = 'red',label='Actual Values')
y_pred = reg.predict(x_input)
plt.plot(x_input,y_pred,label = "Predicted Values On Line of Regression")
plt.legend()
plt.show()
```
