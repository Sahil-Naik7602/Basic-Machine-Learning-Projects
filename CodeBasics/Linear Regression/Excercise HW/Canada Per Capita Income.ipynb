```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df.head(10)
```

```python
%matplotlib inline
plt.figure(figsize=(10,5),dpi = 100)
plt.grid()
plt.scatter(df.year,df['per capita income (US$)'],marker = 'o',color = 'black')
```

```python
model = linear_model.LinearRegression()
x_input = df[['year']]
y_output = df['per capita income (US$)']
model.fit(x_input,y_output)
```

```python
model.predict([[2020]])[0]
```

```python
plt.figure(figsize=(10,5),dpi = 100)
plt.grid()
plt.scatter(df.year,df['per capita income (US$)'],marker = 'o',color = 'black')
y_pred = model.predict(x_input)
plt.plot(x_input,y_output,color = 'yellow')
plt.scatter(x = 2020,y = model.predict([[2020]]))
plt.plot(x_input,y_pred)
plt.show()
```
